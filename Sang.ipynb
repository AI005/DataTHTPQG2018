```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_data=pd.read_csv("2018s.csv")
df_data.head()
```

```python
geograpy_high_score_nationwide=df_data[(df_data.Geography>=8)]["Geography"].count()
civic_education_high_score_nationwide=df_data[(df_data.Civic_Education>=8)]["Civic_Education"].count()
chemistry_high_score_nationwide=df_data[df_data.Chemisty>=8]["Chemisty"].count()
history_high_score_nationwide=df_data[df_data.History>=8]["History"].count()
english_high_score_nationwide=df_data[df_data.English>=8]["English"].count()
literature_high_score_nationwide=df_data[df_data.Literature>=8]["Literature"].count()
biology_high_score_nationwide=df_data[df_data.Biology>=8]["Biology"].count()
math_high_score_nationwide=df_data[df_data.Math>=8]["Math"].count()
physic_high_score_nationwide=df_data[df_data.Physic>=8]["Physic"].count()
a=[geograpy_high_score_nationwide,civic_education_high_score_nationwide,chemistry_high_score_nationwide,history_high_score_nationwide,
  english_high_score_nationwide,literature_high_score_nationwide,biology_high_score_nationwide,math_high_score_nationwide,physic_high_score_nationwide]

count_high_score_nationwide=(df_data[(df_data.Geography>=8)]["Geography"].count()
                  +df_data[df_data.Civic_Education>=8]["Civic_Education"].count()
                 +df_data[df_data.Chemisty>=8]["Chemisty"].count()
                 +df_data[df_data.History>=8]["History"].count()
                 +df_data[df_data.English>=8]["English"].count()
                 +df_data[df_data.Literature>=8]["Literature"].count()
                 +df_data[df_data.Biology>=8]["Biology"].count()
                 +df_data[df_data.Math>=8]["Math"].count()
                 +df_data[df_data.Physic>=8]["Physic"].count())

high_score_proportion_nationwide=count_high_score_nationwide/df_data.ID.count()
```

```python
plt.style.use("fivethirtyeight")
plt.pie(a)
plt.title("Tỷ lệ điểm trên 8 của các môn toàn quốc")
plt.tight_layout()
plt.show
```
