```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_2018 = pd.read_csv('2018.csv', index_col=0)
```

### Fill missing value by -1

```python
df_2018.fillna(-1, inplace=True)
df_2018.isnull().sum()
```

# Cá nhân

# Đánh giá điểm môn Toán và các môn Khoa học tự nhiên

```python
df_hagiang=df_2018[df_2018["city"]==5]
df_hagiang
```

```python
plt.subplots(figsize=(8,6), dpi=100)
sns.distplot([float(i) for i in df_2018["toan"]] , color="dodgerblue", label="Cả nước")
sns.distplot([float(j) for j in df_hagiang["toan"]] , color="orange", label="Hà Giang")
plt.title('Biểu đồ phân phối xác suất điểm Toán của cả nước và Hà Giang')
plt.legend();
```

```python
plt.subplots(figsize=(8,6), dpi=100)
sns.distplot([float(i) for i in df_2018[df_2018["li"] != -1]["li"]] , color="dodgerblue", label="Cả nước")
sns.distplot([float(j) for j in df_hagiang[df_hagiang["li"] != -1]["li"]] , color="orange", label="Hà Giang")
plt.title('Biểu đồ phân phối xác suất điểm Lí của cả nước và Hà Giang')
plt.legend();
```

```python
plt.subplots(figsize=(8,6), dpi=100)
sns.distplot([float(i) for i in df_2018[df_2018["sinh"] != -1]["li"]] , color="dodgerblue", label="Cả nước")
sns.distplot([float(j) for j in df_hagiang[df_hagiang["sinh"] != -1]["li"]] , color="orange", label="Hà Giang")
plt.title('Biểu đồ phân phối xác suất điểm Sinh của cả nước và Hà Giang')
plt.legend();
```

```python
plt.subplots(figsize=(8,6), dpi=100)
sns.distplot([float(i) for i in df_2018[df_2018["hoa"] != -1]["li"]] , color="dodgerblue", label="Cả nước")
sns.distplot([float(j) for j in df_hagiang[df_hagiang["hoa"] != -1]["li"]] , color="orange", label="Hà Giang")
plt.title('Biểu đồ phân phối xác suất điểm Hóa của cả nước và Hà Giang')
plt.legend();
```

```python
plt.subplots(figsize=(8,6), dpi=100)
sns.distplot([float(i) for i in df_2018[df_2018["khtn"] != -1]["li"]] , color="dodgerblue", label="Cả nước")
sns.distplot([float(j) for j in df_hagiang[df_hagiang["khtn"] != -1]["li"]] , color="orange", label="Hà Giang")
plt.title('Biểu đồ phân phối xác suất điểm Toán của cả nước và Hà Giang')
plt.legend();
```

## Nhìn vào các biểu đồ trên ta thấy rằng tỉ lệ điểm từ 8 - 10 của Hà Giang ở các môn Toán, Lí, Sinh, Hóa, KHTN đều cao hơn toàn quốc, điều này cho thấy tỉ lệ các điểm 8-10 đều nằm ở cùng một số học sinh ở tỉnh Hà Giang, đây không thể là sự trùng hợp, chắc chắn có gì bất thường ở đây.
