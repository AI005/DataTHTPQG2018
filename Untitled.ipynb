### Trước tiên ta thử phân tích tổng quan, xem số lượng thí sinh trong top 100 cả nước phân bố ở tỉnh nào

```python
def top_100_comb(subject, data): # top 100 thí sinh cao điểm nhất theo từng tổ hợp môn.
    df = pd.DataFrame({'diem': data[subject].sum(axis=1),
                  'city': df_2018.city.values})
    return df.sort_values(by='diem',ascending = False).head(100)

groups = [
    ['toan', 'li', 'hoa'],
    ['toan', 'hoa', 'sinh'],
    ['van', 'su', 'dia'],
    ['toan', 'li', 'van'],
    ['toan', 'li', 'anh']]

fig, axes = plt.subplots(ncols=1, nrows=5, figsize=(20,30),dpi=200)

plt.subplots_adjust(hspace = 0.8)
for i, ax in zip(range(10), axes.flat):
    top_100_A01 = top_100_comb(groups[i], df_2018)
    sns.countplot(data = top_100_A01, x = 'city',ax=ax)
    ax.set_title(' - '.join(groups[i]).upper())

plt.show()
```

### Ta thử phân tích xem thử phổ điểm của khoa học xã hội và khoa học tự nhiên của Hà Giang so với cả nước thế nào

### Thử nhìn phổ điểm mỗi từng môn để xem rõ hơn

```python
fig, axes = plt.subplots(ncols=3, nrows=3, figsize=(20, 20),dpi=200)
subjects = ['toan', 'li', 'hoa', 'sinh', 'anh','van', 'su', 'dia', 'cd']
count = 0
for i in range(3):
    for j in range(3):
        sns.distplot([float(_) for _ in df_2018[subjects[count]] if _ != -1.00], color="dodgerblue", label="Cả nước", ax=axes[i][j])
        sns.distplot([float(_) for _ in df_HaGiang[subjects[count]] if _ != -1.00], color="orange", label="Hà Giang", ax=axes[i][j])
        axes[i][j].set_title(subjects[i].upper())
        axes[i][j].set_title('Biểu đồ phân phối xác suất điểm {} của cả nước và Hà Giang'.format(subjects[count]))
        axes[i][j].legend()
        count += 1
plt.show()
```

### Tuy nhiên, số lượng thí sinh cao thì tỉnh đó cũng có thể có thành tích cao (Hồ Chí Minh, Hà Nội). Nên ta sẽ xét thêm quan hệ giữa số lượng thí sinh ở tỉnh đó và số lượng thí sinh có thành tích cao ở tỉnh đó
### Ta xây dựng công thức độ đo ``accu``đánh giá dựa vào các tiêu chí phụ thuộc:
###### - `m`: Số lượng thí sinh trên 9 của tỉnh đó
###### - `M`: Tổng thí sinh của tỉnh đó
###### - `n`: Số lượng thí sinh trên 9 của toàn quốc
###### - `N`: Số lượng thí sinh trên toàn quốc
###### * Lưu ý: số lượng thí sinh ở đây là chỉ tính cho một khối thi
##### Công thức: $$accu = \frac{m}{M}*\frac{n}{N}$$

### Thử trực quan hóa "độ đo" từ công thức ở trên giữa các tỉnh

```python
dict_groups = {
    'A':  ['toan', 'li', 'hoa'],
    'A0': ['toan', 'li', 'anh'],
    'B':  ['toan', 'hoa', 'sinh'],
    'C':  ['van', 'su', 'dia'],
    'D':  ['toan', 'li', 'anh']}

def get_score(x):
    if -1 in x.values:
        return -1
    return x.values.mean()

# Tạo thêm các cột điểm 'A', 'B', 'C', 'D'
for key, value in dict_groups.items():
    df_2018[key] = df_2018[value].apply(get_score, axis=1)
    
# Số lượng thí sinh trên 9 của toàn quốc tính các khối A, B, C, D
n_total_g9 = (df_2018[['A', 'B', 'C', 'D']] >= 9).any(axis=1).sum()

# Số lượng thí sinh của Toàn quốc
n_total = df_2018.shape[0]

def get_accuracy(x):
    # Số lượng thí sinh trên 9 của x(tỉnh)
    m = (x[['A', 'B', 'C', 'D', 'A0']] >= 9).any(axis=1).sum()
    
    #Số lượng thí sinh của x(tỉnh)
    M = len(x)
    
    return (m/M*n_total_g9/n_total)*100
df_accu = df_2018.groupby('city').apply(get_accuracy)
plt.figure(figsize=(20,12))
sns.barplot(x=df_accu.index, y=df_accu.values)
plt.show()
```

### Qua đây ta thấy một số tỉnh có khả năng có "vấn đề" là:
####  5   - Hà Giang

####  23 - Hòa Bình
####  30 - Hà Tĩnh
